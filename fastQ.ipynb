```python
from tkinter import *
from tkinter import ttk
from Bio import SeqIO
from urllib.request import urlopen         
from io import TextIOWrapper
from tkinter import filedialog

def openFile():
    global filepath
    filepath = filedialog.askopenfilename(
                                          title="Open file okay?",
                                          filetypes= (("fastq files","*.fastq"),
                                          ("all files",".")))    
def callback():
    file = open(filepath)
    if cmb.get()=="id":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.id)
            i+=1 
    elif cmb.get()=="sequence":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.seq)
            i+=1
    elif cmb.get()=="description":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.description)
            i+=1
    elif cmb.get()=="name":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.name)
            i+=1
    elif cmb.get()=="quilty_control":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.letter_annotations["phred_quality"])
            i+=1 
    elif cmb.get()=="format qual":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.format("qual"))
            i+=1 
    elif cmb.get()=="format fastq":
        i=0
        listbox.delete(0,END)
        for record in SeqIO.parse(file, "fastq"):            
            listbox.insert(i,record.format("fastq"))
            i+=1 
    
    
fastq=Tk()
fastq.title('FASTQ') 
fastq.geometry('800x600')
fontStyle = ("cooper", 17, "bold")
course=["id","sequence","quilty_control","format qual","format fastq","description","name"]
l1=Label(fastq,text="Choose File FASTQ",font = "Helvetica")
l1.grid(row=0, column=0,sticky="NSEW")
button = Button(fastq,text="Open",command=openFile,font = "Helvetica")
button.grid(row=0, column=1,sticky="EW")
cmb=ttk.Combobox(fastq,values=course,width=30,font = "Helvetica")
cmb.grid(row=1, column=1,sticky="EW")
cmb.current(0)
l2=Label(fastq,text="Choose Feature",font = "Helvetica")
l2.grid(row=1, column=0,sticky="NSEW")
btn=Button(fastq,text="Click Here",command=callback,font = "Helvetica")
btn.grid(row=2, column=1,sticky="EW")
labelframe = LabelFrame(fastq, text="View",font = "Helvetica")
labelframe.grid(column=0, row=3, sticky="NSEW")

Grid.rowconfigure(fastq,0,weight=1)
Grid.rowconfigure(fastq,2,weight=1)
Grid.rowconfigure(fastq,3,weight=1)
Grid.columnconfigure(fastq,0,weight=1)
Grid.columnconfigure(fastq,1,weight=1)
Grid.columnconfigure(fastq,3,weight=1)
Grid.rowconfigure(fastq,1,weight=1)
listbox = Listbox(labelframe,height = 25, 
                  width = 70,
                  activestyle = 'dotbox')
listbox.pack(side = LEFT, fill = BOTH)
  
scrollbary = Scrollbar(labelframe)
scrollbary.pack(side = RIGHT, fill = Y)
scrollbarx = Scrollbar(labelframe,orient=HORIZONTAL)
scrollbarx.pack(side = BOTTOM, fill = X)
listbox.config(yscrollcommand = scrollbary.set)
scrollbary.config(command = listbox.yview)


listbox.config(xscrollcommand = scrollbarx.set)

scrollbarx.config(command = listbox.xview)
fastq.mainloop()
```
