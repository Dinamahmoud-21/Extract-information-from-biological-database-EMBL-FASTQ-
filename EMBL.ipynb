```python
from tkinter import *
from tkinter import ttk
from Bio import SeqIO
from urllib.request import urlopen         
from io import TextIOWrapper 
def Buclick():
    global record
    global dic
    handle =TextIOWrapper(urlopen(entry.get()))
    record=SeqIO.read(handle, "embl")
    dic={}
    if record.features:
        for feature in record.features:
            if feature.type == "CDS":
                dic["type_feature"]=feature.type
                dic["location_exon"]=feature.location
                dic["codon_start"]=feature.qualifiers["codon_start"]
                dic["db_xref_exon"]=feature.qualifiers["db_xref"]
                dic["gene"]=feature.qualifiers["gene"]
                dic["note"]=feature.qualifiers["note"]
                dic["locus_tag"]=feature.qualifiers["locus_tag"]
                dic["product"]=feature.qualifiers["product"]
                dic["protein_id"]=feature.qualifiers["protein_id"]
                dic["translation"]=feature.qualifiers["translation"]

def callback():
    if cmb.get()=="id":
        listbox.delete(0,END)
        listbox.insert(1,record.id)
    elif cmb.get()=="name":
        listbox.delete(0,END)
        listbox.insert(1,record.name)
    elif cmb.get()=="description":
        listbox.delete(0,END)
        listbox.insert(1,record.description)
    elif cmb.get()=="dbxrefs":
        listbox.delete(0,END)
        listbox.insert(1,record.dbxrefs)
    elif cmb.get()=="annotations":
        listbox.delete(0,END)
        listbox.insert(1,record.annotations)
    elif cmb.get()=="sequence":
        listbox.delete(0,END)
        listbox.insert(1,record.seq)
    elif cmb.get()=="type feature":
        listbox.delete(0,END)
        listbox.insert(1,dic["type_feature"])
    elif cmb.get()=="location_exon":
        listbox.delete(0,END)
        listbox.insert(1,dic["location_exon"])
    elif cmb.get()=="codon_start":
        listbox.delete(0,END)
        listbox.insert(1,dic["codon_start"])
    elif cmb.get()=="db_xref_exon":
        listbox.delete(0,END)
        listbox.insert(1,dic["db_xref_exon"])
    elif cmb.get()=="locus_tag":
        listbox.delete(0,END)
        listbox.insert(1,dic["locus_tag"])
    elif cmb.get()=="note":
        listbox.delete(0,END)
        listbox.insert(1,dic["note"])
    elif cmb.get()=="product":
        listbox.delete(0,END)
        listbox.insert(1,dic["product"])
    elif cmb.get()=="gene":
        listbox.delete(0,END)
        listbox.insert(1,dic["gene"])
    elif cmb.get()=="protein_id":
        listbox.delete(0,END)
        listbox.insert(1,dic["protein_id"])
    elif cmb.get()=="translation":
        listbox.delete(0,END)
        listbox.insert(1,dic["translation"])
    
window=Tk()
window.title('EMBL') 
window.geometry('900x700')
fontStyle = ("cooper", 17, "bold")
features=["id","name","description","dbxrefs","annotations","sequence","type feature","location_exon","codon_start","db_xref_exon","gene","note","locus_tag","product","protein_id","translation"]

l1=Label(window,text="Enter Link",font = "Helvetica")
l1.grid(column=0, row=0,sticky="NSEW")

entry=Entry(window,font = "Helvetica")
entry.grid(column=1, row=0,sticky="EW")

button1 =Button(window,text='GO',command=Buclick,font = "Helvetica")
button1.grid(column=3, row=0,sticky="EW")

l2=Label(window,text="Choose Feature",font = "Helvetica")
l2.grid(column=0, row=1,sticky="NSEW")

cmb=ttk.Combobox(window,values=features,width=30,font = "Helvetica")
cmb.grid(column=1, row=1,sticky="EW")
cmb.current(0)

btn=Button(window,text="OK",command=callback,font = "Helvetica")
btn.grid(column=3, row=1,sticky="EW")

labelframe = LabelFrame(window, text="View",font = "Helvetica")
labelframe.grid(column=0, row=2, sticky="SNEW")

Grid.rowconfigure(window,0,weight=1)
Grid.rowconfigure(window,2,weight=1)
Grid.rowconfigure(window,3,weight=1)
Grid.columnconfigure(window,0,weight=1)
Grid.columnconfigure(window,2,weight=1)
Grid.columnconfigure(window,1,weight=1)
Grid.columnconfigure(window,3,weight=1)
Grid.rowconfigure(window,1,weight=1)
listbox = Listbox(labelframe,height = 20, 
                  width = 50,
                  activestyle = 'dotbox', 
                  font = "Helvetica")
listbox.pack(side = LEFT, fill = BOTH)
  
scrollbary = Scrollbar(labelframe)
scrollbary.pack(side = RIGHT, fill = Y)
scrollbarx = Scrollbar(labelframe,orient=HORIZONTAL)
scrollbarx.pack(side = BOTTOM, fill = X)
listbox.config(yscrollcommand = scrollbary.set)
scrollbary.config(command = listbox.yview)


listbox.config(xscrollcommand = scrollbarx.set)

scrollbarx.config(command = listbox.xview)
window.mainloop()

#https://www.ebi.ac.uk/ena/browser/api/embl/EAL24519.1?lineLimit=1000
```
